```python
data = pd.read_csv('dataset_184_covertype.csv')
data.head()
```

```python
x = data.drop("class", axis=1)
y = data["class"]
```

```python
error = 0.0
for i, sample in enumerate(x.head().iterrows()):
    model = knn.KNN(x.head().drop(i), y.drop(i), k=1)
    print(model.margin(sample[1], y[i]))
```
