```python
import math


def manhattan(v1, v2):
    s = 0
    for i in range(len(v1)):
        s += abs(v1[i] - v2[i])
    return s


def euclidean(v1, v2):
    s = 0
    for i in range(len(v1)):
        s += (v1[i] - v2[i]) ** 2
    return math.sqrt(s)


def chebyshev(v1, v2):
    s = float('-inf')
    for i in range(len(v1)):
        s = max(s, abs(v1[i] - v2[i]))
    return s


def uniform(d):
    if not -1 < d < 1:
        return 0
    return 0.5


def triangular(d):
    if not -1 < d < 1:
        return 0
    return 1 - abs(d)


def epanechnikov(d):
    if not -1 < d < 1:
        return 0
    return 0.75 * (1 - d ** 2)


def quartic(d):
    if not -1 < d < 1:
        return 0
    return 15 / 16 * (1 - d ** 2) ** 2


def triweight(d):
    if not -1 < d < 1:
        return 0
    return 35 / 32 * (1 - d ** 2) ** 3


def tricube(d):
    if not -1 < d < 1:
        return 0
    return 70 / 81 * (1 - abs(d) ** 3) ** 3


def gaussian(d):
    return 1 / math.sqrt(2 * math.pi) * math.exp(-0.5 * d ** 2)


def cosine(d):
    if not -1 < d < 1:
        return 0
    return math.pi / 4 * math.cos(math.pi / 2 * d)


def logistic(d):
    return 1 / (math.exp(d) + 2 + math.exp(-d))


def sigmoid(d):
    return 2 / math.pi / (math.exp(d) + math.exp(-d))
```
